```python
import cv2
import os

# Define paths
video_name = "download_video.mp4"
video_path = r"D:\Waqar\My-Learning-Projects\Jupyter-Notebook"
save_path = r"D:\\Waqar\My-Learning-Projects\Jupyter-Notebook\possitive-video-frames-bmp"

# Ensure the save directory exists
os.makedirs(save_path, exist_ok=True)

# Load the video
video_file = os.path.join(video_path, video_name)
cap = cv2.VideoCapture(video_file)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

frame_count = 0

# Loop through video frames
while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        break
    
    # Save the frame as BMP image
    frame_filename = os.path.join(save_path, f"frame_{frame_count:04d}.bmp")
    cv2.imwrite(frame_filename, frame)
    
    print(f"Saved {frame_filename}")
    frame_count += 1

# Release the video capture object
cap.release()
print("Conversion complete!")



# rawdata/frame_0000.bmp 1 229 275 84 90
# rawdata/frame_0001.bmp 1 222 284 84 91
# rawdata/frame_0002.bmp 1 219 283 81 102
# rawdata/frame_0003.bmp 1 219 291 76 105
# rawdata/frame_0004.bmp 1 204 297 83 112
# rawdata/frame_0005.bmp 1 194 296 99 120
# rawdata/frame_0006.bmp 1 193 305 81 121
# rawdata/frame_0007.bmp 1 183 305 104 128
# rawdata/frame_0008.bmp 1 166 315 118 147
# rawdata/frame_0009.bmp 1 169 313 107 154
# rawdata/frame_0010.bmp 1 141 335 133 144
# rawdata/frame_0011.bmp 1 148 331 111 166
# rawdata/frame_0012.bmp 1 123 335 137 176
# rawdata/frame_0013.bmp 1 116 348 132 172
# rawdata/frame_0014.bmp 1 112 354 133 179
# rawdata/frame_0015.bmp 1 90 366 152 186
# rawdata/frame_0016.bmp 1 94 382 138 190
# rawdata/frame_0017.bmp 1 79 392 145 196
# rawdata/frame_0018.bmp 1 57 400 155 206
# rawdata/frame_0019.bmp 1 44 417 161 210
# rawdata/frame_0020.bmp 1 6 448 195 200
# rawdata/frame_0021.bmp 1 2 443 186 227

```
